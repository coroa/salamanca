```python
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('seaborn-deep')

%matplotlib inline
```

```python
xmax = 5
linewidth = 5

x = np.linspace(0, xmax)
for factor in [-.5, -0.25, 0, 0.25, .5]:
    y = np.sin(x) + factor * x
    plt.plot(x, y, linewidth=linewidth)

x = np.linspace(-1, 0)
plt.plot(x, np.sin(x), linewidth=linewidth)

plt.xlim(-1, xmax)
plt.axis('off')
plt.tight_layout()
```

```python
from scipy.stats import lognorm
```

```python
x = np.linspace(0, 1)
shapes = 1 + np.array(range(1, 5)) / 7.
for s in shapes[::-1]:
    _s = 0.5
    y = lognorm.pdf(x * 2, 1.25, scale=s)
    plt.plot(x, y / 5, linewidth='2')
plt.title('Representative Income Distributions')
plt.xlabel('Income (relative)')
plt.ylabel('Population (relative)')
plt.savefig('income.png')
```
